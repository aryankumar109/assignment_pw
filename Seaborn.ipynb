Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful data visualization library built on top of Matplotlib. Here are five common types of plots you can create using Seaborn, along with their typical use cases:

1. **Scatter Plot (`sns.scatterplot`):**
   - **Use**: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a data point, making it suitable for exploring the distribution of data points and identifying patterns or trends.

2. **Histogram (`sns.histplot` or `sns.distplot`):**
   - **Use**: Histograms are used to visualize the distribution of a single variable. They help you understand the frequency or density of data points within different bins or intervals. It's useful for assessing data's central tendency, spread, and skewness.

3. **Bar Plot (`sns.barplot`):**
   - **Use**: Bar plots are used to display and compare the values of categorical variables. They represent data using vertical or horizontal bars, making it easy to compare different categories. Bar plots are useful for visualizing counts, proportions, or aggregations of data within categories.

4. **Box Plot (`sns.boxplot`):**
   - **Use**: Box plots, also known as box-and-whisker plots, are used to visualize the distribution and spread of continuous variables, especially in comparing multiple categories or groups. They show the median, quartiles, and potential outliers in the data.

5. **Heatmap (`sns.heatmap`):**
   - **Use**: Heatmaps are used to display the relationships between two categorical variables or to visualize the correlation matrix of numerical variables. They use color intensity to represent the strength of relationships or correlations, making it easier to identify patterns and associations in complex datasets.

These are just a few examples of the types of plots you can create with Seaborn. Seaborn provides a wide range of other plots and customization options to suit various data visualization needs.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn's lineplot function
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Use the lineplot function to create the line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Add labels and a title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of Signal vs. Timepoint for Different Events and Regions")

# Show the legend
plt.legend(title="Event")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create subplots for two box plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for 'age' vs. 'pclass'
sns.boxplot(data=titanic_data, x='pclass', y='age', ax=axes[0])
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')
axes[0].set_title('Box Plot of Age vs. Pclass')

# Box plot for 'fare' vs. 'pclass'
sns.boxplot(data=titanic_data, x='pclass', y='fare', ax=axes[1])
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')
axes[1].set_title('Box Plot of Fare vs. Pclass')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Use Seaborn's histplot to create the histogram with hue
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)

# Add labels and a title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices with Cut Variations')

# Show the legend
plt.legend(title='Cut')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for the 'species' column
sns.set(style="ticks")
sns.pairplot(iris_data, hue="species")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap using Seaborn's heatmap function
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap (Flights Dataset)')

# Show the plot
plt.show()
```
