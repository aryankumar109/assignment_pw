A probability density function (PDF) is a fundamental concept in probability and statistics. It describes the probability distribution of a continuous random variable. In other words, it tells you how the probabilities are distributed across the possible values of that variable.

Different types of continuous probability distributions (e.g., normal distribution, exponential distribution, uniform distribution) have specific PDFs that describe how their probabilities are distributed over their respective ranges. These PDFs are essential in various statistical analyses and applications, including calculating probabilities, expected values, and performing hypothesis testing.

There are several types of probability distributions in statistics. These distributions describe how the probabilities are distributed across the possible values of a random variable. The choice of distribution depends on the characteristics of the data or the random process being modeled. Here are some common types of probability distributions:

    Uniform Distribution:
        All values within a given range are equally likely.
        Example: Rolling a fair six-sided die, where each outcome has a 1/6 probability.

    Normal Distribution (Gaussian Distribution):
        Often referred to as the bell curve.
        Symmetrical, with a peak at the mean and tails that extend to infinity.
        Many real-world phenomena approximate a normal distribution.
        Example: Heights or weights of a population.

    Binomial Distribution:
        Describes the number of successes (e.g., heads in coin flips) in a fixed number of independent Bernoulli trials (experiments with two possible outcomes).
        Parameters include the number of trials and the probability of success.
        Example: Probability of getting 3 heads in 5 coin flips.

    Poisson Distribution:
        Models the number of events that occur within a fixed interval of time or space.
        Applicable when events occur randomly and independently at an average rate.
        Example: The number of customer arrivals at a store in an hour.

    Exponential Distribution:
        Describes the time between events in a Poisson process.
        Often used to model waiting times between events.
        Example: The time between arrivals of customers at a store (if arrivals follow a Poisson process).

    Bernoulli Distribution:
        A simple distribution with two possible outcomes, usually labeled as success and failure.
        Parameterized by the probability of success (p).
        Example: Flipping a coin (success = heads, failure = tails).

```python
from scipy.stats import norm

def calculate_normal_pdf(x, mean, std_dev):
  
    pdf = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf

# Example usage:
mean = 0
std_dev = 1
x = 1
pdf_value = calculate_normal_pdf(x, mean, std_dev)
print(f"PDF at x={x} is {pdf_value:.4f}")
```

The binomial distribution is a discrete probability distribution that is widely used in statistics and probability theory. It is characterized by the following properties:

1. Fixed Number of Trials (n): The binomial distribution models a fixed number of independent and identical trials or experiments, denoted by 'n.' In other words, you perform a specific number of trials or experiments under the same conditions.

2. Two Possible Outcomes: Each trial has only two possible outcomes, often referred to as "success" and "failure." These outcomes are mutually exclusive and exhaustive.

3. Constant Probability of Success (p): The probability of success (often denoted as 'p') remains constant from trial to trial. In other words, the probability of success does not change as you conduct more trials.

4. Independence: The trials are assumed to be independent, meaning that the outcome of one trial does not affect the outcomes of subsequent trials.

5. Discreteness: The random variable X, which represents the number of successes in 'n' trials, takes on discrete values (usually non-negative integers) from 0 to n.

The probability mass function (PMF) of the binomial distribution is given by:

P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)

Where:
- P(X = k) is the probability of having 'k' successes in 'n' trials.
- C(n, k) is the binomial coefficient, also known as "n choose k," which is equal to (n! / (k! * (n - k)!).
- p is the probability of success in a single trial.
- (1 - p) is the probability of failure in a single trial.

Two common examples of events where the binomial distribution can be applied are:

1. Coin Flipping: When flipping a fair coin (where the probability of getting heads, for example, is p = 0.5), and you want to find the probability of getting a specific number of heads in a fixed number of flips (n), the binomial distribution can be used.

2. Product Defects: In a manufacturing process, if a product has a known probability of being defective (p), and you want to calculate the probability of a certain number of defective products in a sample of a fixed size (n), the binomial distribution can be used.

These are just a couple of examples, but the binomial distribution is applicable in various real-world scenarios where you have a fixed number of trials with two possible outcomes and a constant probability of success.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the binomial distribution
n = 1000  # Sample size
p = 0.4   # Probability of success

# Generate the random sample
random_sample = np.random.binomial(n, p, size=1000)

# Create a histogram of the results
plt.hist(random_sample, bins=range(0, n + 2), density=True, alpha=0.75, color='b', edgecolor='black')

# Customize the plot labels and title
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution with n=1000 and p=0.4')

# Display the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import math

def poisson_cdf(mean, k):
 
    if mean < 0 or k < 0:
        raise ValueError("Both mean and k must be non-negative values.")
    
    cdf = 0.0
    for i in range(k + 1):
        cdf += math.exp(-mean) * (mean**i) / math.factorial(i)
    
    return cdf

# Example usage:
mean = 3.0
k = 2
cdf = poisson_cdf(mean, k)
print(f"P(X ≤ {k}) = {cdf:.4f}")
```

The key difference between the Binomial and Poisson distributions is in the nature of the random events they model. The Binomial distribution is used for a fixed number of independent trials with two outcomes (success and failure), whereas the Poisson distribution models events occurring in a fixed interval of time or space with a known average rate, without a fixed number of trials.

```python
import numpy as np

# Set the parameters
mean = 5
sample_size = 1000

# Generate the random sample
random_sample = np.random.poisson(mean, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

    Mean (μ):
        In a Binomial distribution with parameters n (number of trials) and p (probability of success in each trial), the mean (μ) is 
        given by μ = n * p.
        In a Poisson distribution with a mean λ (average rate of events in a fixed interval), the mean (μ) is also equal to λ.

    Variance (σ^2):
        In a Binomial distribution with parameters n and p, the variance (σ^2) is given by σ^2 = n * p * (1 - p).
        In a Poisson distribution with a mean λ, the variance (σ^2) is also equal to λ.
