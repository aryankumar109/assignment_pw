Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual shapes or markers that you can use to represent data points on a plot. Glyphs allow you to customize the appearance of your data by specifying various attributes such as size, color, and shape. Bokeh provides a wide range of built-in glyphs, including circles, squares, lines, bars, and more.

To add glyphs to a Bokeh plot, you typically follow these steps:

Create a figure:Use the figure function to create a blank figure where you'll add your glyphs.

Define your data: Prepare your data in a format that can be used by Bokeh. This could be a Pandas DataFrame, NumPy arrays, or Bokeh's ColumnDataSource object.

Add glyphs: Use the appropriate glyph method for your data type (e.g., circle, line, rect, etc.) to add glyphs to the figure. Specify the data source, and customize the glyph's attributes such as size, color, and shape.

Customize the plot: You can further customize the plot by setting attributes for the axes, labels, titles, and legends.

Show the plot: Finally, use the show function to display the plot in a web browser or as an output file.

```python
from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add glyphs (circles) to the figure
p.circle(x, y, size=10, color='blue', legend_label="Data Points")

# Customize the plot
p.title.text_font_size = "16pt"
p.legend.label_text_font_size = "12pt"

# Show the plot
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


```python
from bokeh.plotting import figure, show

# Create a figure
p = figure(
    title="Customized Bokeh Plot",
    x_axis_label="X-axis Label",
    y_axis_label="Y-axis Label",

    background_fill_color="lightgray",
    border_fill_color="white",
)

# Customize axes
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_color = "green"
p.xaxis.axis_label_text_font_size = "14pt"
p.yaxis.axis_label_text_font_size = "14pt"
p.xaxis.major_label_text_color = "navy"
p.yaxis.major_label_text_color = "darkgreen"

# Customize legend
p.legend.title = "Legend Title"
p.legend.location = "top_right"
p.legend.background_fill_color = "lightyellow"
p.legend.border_line_color = "orange"

# Customize grid lines
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"
p.xgrid.grid_line_dash = [6, 4]
p.ygrid.grid_line_dash = [6, 4]

# Add data (glyphs)
p.circle([1, 2, 3], [4, 5, 6], size=10, color="blue", legend_label="Data Series 1")
p.square([2, 3, 4], [6, 5, 4], size=10, color="green", legend_label="Data Series 2")

# Show the plot
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a feature of the Bokeh library that allows you to create interactive web applications with dynamic and real-time updating plots. With the Bokeh server, you can build data-driven applications that respond to user interactions, streaming data, or other external events.

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource, Slider

# Create a data source
source = ColumnDataSource(data={'x': [], 'y': []})

# Create a plot
plot = figure(title="Real-Time Updating Plot")
line = plot.line('x', 'y', source=source)

# Define a callback function to update the plot
def update(attr, old, new):
    # Generate new data points
    x = list(range(0, slider.value))
    y = [i**2 for i in x]

    # Update the data source
    source.data = {'x': x, 'y': y}

# Create a slider widget
slider = Slider(start=1, end=10, step=1, value=1, title="Number of Points")
slider.on_change('value', update)

# Add the plot and slider to the current document
curdoc().add_root(plot)
curdoc().add_root(slider)

```
